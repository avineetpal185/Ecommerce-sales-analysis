#CONVERTING DATE COLUMNS !

```python
#like above we have givent the date and time in the string , spo convert -> number !

#I amsying this column in table -> cahneg into datetime !
data['Order Date'] = pd.to_datetime(data['Order Date'])
#for ship Date also !
data['Ship Mode'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
#Now if you see our Order-Date  <- type chaneg from object to datetime !
```

#Now i make 3 new column for better anlsis ! like ->year,month,order
#year -> year data !
#month->month data collect !
#order ->date  (when order departcher and receved (it on time ? or delay !))

```python
#Let create-

#1) col =below for month  ->(and i extratc month from->order_date column), i want only month !
data['Order Month'] = data['Order Date'].dt.month
# -> .dt is a datetime accessor in Pandas.
#-> An accessor is something that lets you access special features of a column.

#2)col = now i making for year -> i want only year !
data['Order Year'] = data['Order Date'].dt.year

#3)- For the week !
data['Order Day of Week'] = data['Order Date'].dt.dayofweek

#When the column not exit pandas create the new column , but if already exist they update inside the same column !
```

Q6)-Analyze the sales and Profit by customer segment ?

```python
#we find the sales and profit , now they are saying sales and profit find both together for customer -segment type !
#cusotmer-segment liek -> buissmen ,Coporate customer, Consumer etc !
```

```python
data.head()

#we will take 3 column ->customer segement, proft , sales 
```

```python
#Important below !

#Taking column -> segmenet , aggrigate for 2 column ->Sales and profit (there sum)
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

#bar graph!
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],  
                    y =sales_profit_by_segment['Sales'],
                    name ='Sales',
                    marker_color = color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                    y =sales_profit_by_segment['Profit'],
                    name ='Profit',
                    marker_color = color_palette[1]))


fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                 xaxis_title='Customer Segment',yaxis_title='Amount')
fig.show()

#addtrace -> add_trace() means → add one chart (one data layer) inside the graph. 

#If you want multiple things in one graph, you use add_trace(). , we can create mutiple line or create mutiple pie in this ! by add_trace !

#I can also control the gap between the bar 
#fig.update_layout(
#    bargap = 0.7
#     )
```

Q7-Analyze the sales to Profit ratio !

```python
#tell how much sales-> there ratio !   
#so -> sales/profit !
```

```python
#below doing grup by ->Segment column , Sales and Prifut column (sum) !
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

                                                        #below dividing(sales/profit)for ratio !
sales_profit_by_segment['Sales_to_Profit_ratio'] = sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_ratio']])
```

Q-WHICH CATEGORY HAS THE HIGHEST MARGIN PROFIT ?

```python
pr_mar = data.groupby('Category').agg({
    'Sales':'sum',
    'Profit':'sum'
})
pr_mar
pr_mar['Profit_margin'] = pr_mar['Profit']/pr_mar['Sales']
pr_mar

#Now let find the highest amrgin
result = pr_mar.sort_values(by= 'Profit_margin',ascending=False)
result
```

```python
result = result.reset_index()
fig1 = px.bar(result,
              x = 'Category',
              y = 'Profit_margin',
              title='HIGHEST MARGIN'
    
)
fig1.show()
```
